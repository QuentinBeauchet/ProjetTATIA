Le fichier qui nous permet de recuperer nos données du site Allociné est le fichier dataMaker.py

La fonction loadPages prend en parametre l'id d'un film et son nom (optionnel) puis lis tous les commentaires utilisateurs du film et les sauvegarde avec leurs notes dans un fichier data/id.csv

La fonction renvoit la note calculée en fonction des commentaires recuperés, car la note presente sur le site prend aussi en compte les avis sans commentaires, et le nombre de commentaires total recuperés.

Nous vous avons fournis les fichiers csv deja telechargées car pour les 40 films cela prend jusqu'a 20min.

Pour determiner la classe selon le commentaire nous nettoyons d'abord celui ci en enlevant le ponctuation et les mots de liaisons non negatifs grace a la fonction cleanup.

Puis nous faisons un classification_report de la librairie sklearn ce qui nous permet d'avoir la precision de la matrice de confusion des classes.

Au final la fonction renvoit le dictionnaie de la classification_report et un Serie contenant le nombre de commentaires de chaque classe estimées.

```python
import analyser

analyser.analyseMovie(208104)
```

Pour analyser plus rapidement un nombre important de films et pour faire un estimation plus generalle de l'efficacité de notre estimation nous utilisons la fonction analyseAll.

Celle ci va appeler analyseMovie pour chaque film du fichier data/data.csv puis determine une note selon les rapports entre le nombre de commentaires Positive, Neutral et Negative.
Elle determine aussi la classe du film selon deux methodes differentes, l'une selon la note estimé precedement et l'autre selon le rapport entre Commentaires Positive et Negative.
Elle stocke aussi la precision de l'estimation de chaque classe et la precision globale et pondérée.

Puis elle sauvegarde ces données dans un fichier data/analysis.csv et renvoit cette DataFrame.

```python
analyser.analyseAll()
```

Pour savoir l'efficacité de nos estimation il faut appeler la fonction printResults sur la DataFrame precedment generée.

Ces données sont sauvegardées dans les fichiers data/statsComment.csv et data/statsClasse.csv
